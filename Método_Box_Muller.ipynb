```python
import numpy as np
import matplotlib.pyplot as plt

def box_muller(n):
    # Generar n números aleatorios uniformes en (0,1)
    u1 = np.random.rand(n//2)
    u2 = np.random.rand(n//2)

    # Aplicar la transformación de Box-Muller
    z1 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    z2 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)

    # Combinar los dos conjuntos de números normales
    return np.concatenate((z1, z2))

# Número de muestras
n = 10000

# Generar muestras normales estándar
samples = box_muller(n)

# Calcular media y desviación estándar
media = np.mean(samples)
desviacion = np.std(samples)

# Mostrar resultados
print(f"Media de las muestras: {media:.4f}")
print(f"Desviación estándar de las muestras: {desviacion:.4f}")

# Dibujar histograma
plt.hist(samples, bins=50, density=True, color='skyblue', edgecolor='black')
plt.title("Histograma de números normales estándar (Box-Muller)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```
