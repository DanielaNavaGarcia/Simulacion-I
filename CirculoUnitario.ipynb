```python
import random
import matplotlib.pyplot as plt
import numpy as np
# Listas para almacenar las coordenadas de los dardos
x_dentro = []
y_dentro = []
x_fuera = []
y_fuera = []
```

```python

# Parámetros de la simulación
num_dardos = 72 # Número de dardos a lanzar
dardos_dentro = 0  # Contador de dardos que caen dentro del cuarto de círculo
```

```python
def simular_dardo_cuarto():
  """Simula el lanzamiento de un dardo en un cuarto de círculo."""
  x = random.uniform(0, 1)
  y = random.uniform(0, 1)
  if x**2 + y**2 <= 1 and x >= 0 and y >= 0:
    x_dentro.append(x)
    y_dentro.append(y)
    return True  # El dardo cae dentro del cuarto de círculo
  else:
    x_fuera.append(x)
    y_fuera.append(y)
    return False  # El dardo cae fuera del cuarto de círculo

# Simulación
for _ in range(num_dardos):
  if simular_dardo_cuarto():
    dardos_dentro += 1



```

```python
#  Resultados

print(f"Número de dardos lanzados: {num_dardos}")
print(f"Número de dardos dentro del cuarto de círculo: {dardos_dentro}")

# for i in range(len(x_dentro)):
#   print(f"Coordenadas del dardo dentro: ({x_dentro[i]}, {y_dentro[i]})")

# for i in range(len(x_fuera)):
#   print(f"Coordenadas del dardo fuera: ({x_fuera[i]}, {y_fuera[i]})")
```

```python
# Gráfica
plt.figure(figsize=(6, 6))  # Ajusta el tamaño de la figura

# Dibuja el cuarto de círculo
theta = np.linspace(0, np.pi/2, 100)
x_circulo = np.cos(theta)
y_circulo = np.sin(theta)
plt.plot(x_circulo, y_circulo, color='black', linewidth=2)

# Dibuja los puntos
plt.scatter(x_dentro, y_dentro, color='blue', s=5, label='Dardos dentro')
plt.scatter(x_fuera, y_fuera, color='red', s=5, label='Dardos fuera')

# Configuraciones adicionales
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simulación de")
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.legend()
plt.grid(True)

plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np
# Listas para almacenar las coordenadas de los dardos
x_dentro = []
y_dentro = []
x_fuera = []
y_fuera = []


# Parámetros de la simulación
num_dardos = 25 # Número de dardos a lanzar
dardos_dentro = 0  # Contador de dardos que caen dentro del cuarto de círculo

def simular_dardo_cuarto():
  """Simula el lanzamiento de un dardo en un cuarto de círculo."""
  x = random.uniform(-1, 1)
  y = random.uniform(-1, 1)
  if x**2 + y**2 <= 1 : #and x >= 0 and y >= 0 condiciones 1/4
    x_dentro.append(x)
    y_dentro.append(y)
    return True  # El dardo cae dentro del cuarto de círculo
  else:
    x_fuera.append(x)
    y_fuera.append(y)
    return False  # El dardo cae fuera del cuarto de círculo

# Simulación
for _ in range(num_dardos):
  if simular_dardo_cuarto():
    dardos_dentro += 1

#  Resultados

print(f"Número de dardos lanzados: {num_dardos}")
print(f"Número de dardos dentro del cuarto de círculo: {dardos_dentro}")

# for i in range(len(x_dentro)):
#   print(f"Coordenadas del dardo dentro: ({x_dentro[i]}, {y_dentro[i]})")

# for i in range(len(x_fuera)):
#   print(f"Coordenadas del dardo fuera: ({x_fuera[i]}, {y_fuera[i]})")

# Gráfica
plt.figure(figsize=(6, 6))  # Ajusta el tamaño de la figura

# Dibuja el cuarto de círculo
theta = np.linspace(0, 2* np.pi, 100)
x_circulo = np.cos(theta)
y_circulo = np.sin(theta)
plt.plot(x_circulo, y_circulo, color='pink', linewidth=2)

# Dibuja los puntos
plt.scatter(x_dentro, y_dentro, color='purple', s=5, label='Dardos dentro')
plt.scatter(x_fuera, y_fuera, color='orange', s=5, label='Dardos fuera')

# Configuraciones adicionales
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simulación de lanzamientos de dardos")
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.legend()
plt.grid(True)

plt.show()
```
