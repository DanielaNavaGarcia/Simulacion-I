```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal
mu = 10           # media
sigma = 2         # desviación estándar
n_muestras = 10000

# Método de las 12 uniformes
uniformes = np.random.uniform(0, 1, (n_muestras, 12))
normales_estandar = np.sum(uniformes, axis=1) - 6  # media 0, varianza 1
normales = mu + sigma * normales_estandar          # escalamos a N(mu, sigma^2)

# Resultados estadísticos
media_muestral = np.mean(normales)
desv_muestral = np.std(normales)

print(f"Media muestral: {media_muestral:.4f}")
print(f"Desviación estándar muestral: {desv_muestral:.4f}")

# Histograma
plt.hist(normales, bins=50, density=True, alpha=0.6, color='pink', edgecolor='black')
plt.title("Histograma de muestras generadas (Método de las 12 uniformes)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```
