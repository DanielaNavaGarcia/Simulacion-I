```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import randint
```

```python
alumnos = pd.read_csv("Alumnos.csv", encoding="latin-1")
print(alumnos)
```

```python
alumnos ['Nombre Completo'] = alumnos['Nombre']+ ' '+alumnos['Apellido']
print(alumnos)
```

```python
np.random.seed(42)
```

```python
datos = {
    "Nombre Completo": alumnos['Nombre Completo'],
    "Edad": np.random.randint(18,27, size=27),
    "Calificación": np.random.normal(7.5, 1.0, size=27).round(1),
    "Asistencia (%)": np.random.uniform(70, 100, size=27).round(2)
}
df = pd.DataFrame(datos)
print(df)
```

```python
plt.hist(df['Edad'], bins=6, edgecolor='black', color='pink')
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de edades de los alumnos")
plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df["Nombre Completo"], df["Asistencia (%)"], color='pink', edgecolor='black')
plt.xlabel("Alumno")
plt.ylabel("Asistencia (%)")
plt.title("Porcentaje de asistencia por alumno")
plt.xticks(rotation=90)
plt.ylim(50,100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df["Edad"], df["Asistencia (%)"], color='pink')
plt.xlabel("Edad")
plt.ylabel("Asistencia (%)")
plt.title("Relacion entre Edad y Porcentaje de Asistencis")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df["Nombre Completo"], df["Calificación"], color='pink', edgecolor='black')
plt.xlabel("Alumno")
plt.ylabel("Calificación")
plt.title("Calificaciones de los alumnos")
plt.xticks(rotation=90)
plt.ylim(0,10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
