```python
import pandas as pd
import tabulate as tb
import matplotlib.pyplot as plt
import numpy as np

# Definicion de variables

lista =[]

a = 10**45+1;
c = 103;
m = 10**4 - 1;
x_0 = 5;
lista = [];

# Funcion Generadora
def generador(a,c,m,x_0):
        x= (a*x_0 + c) % m;
        return x;

# Imprime las n iteraciones de la funcion
for i in range(m):
    x_0 = generador(a,c,m,x_0);
    lista.append(x_0/m);

print(lista);

# plt.plot(lista);
# Histograma
plt.hist(lista);
plt.ylim(0, 1);
plt.show();

print(np.mean(lista));
print(np.var(lista));

# Grafica de probabilidad acumulada
lista_sort = np.sort(lista);
print(lista_sort);

y = np.arange(1, m + 1)/m;
plt.plot(lista_sort, y);
plt.show();









```
